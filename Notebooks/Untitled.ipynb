```python
import pandas as pd
import numpy as np
```

```python
data_file='data\\Final_Berg JBS 2013 Supplemental Table 3_For SVM14Dec2017.xlsx'
mechanism_file='data\\Final_Berg JBS 2013 Supplemental Table 3_For SVM14Dec2017 - Mechanisms.xlsx'
    
data = pd.read_excel(data_file)


mech = pd.read_excel(mechanism_file)
# merge mechanism 
data = pd.merge(left=data, right=mech, how='left')


data['Agent'] = [s.split(',')[2] for s in data.iloc[:, 0]]
data['Concentration']  = [s.split(',')[4] for s in data.iloc[:, 0]]

###############################
# get certain columns as index
#index = ['agent', 'conc', 'mech', 'prof']
index = ['mech']

if isinstance(index, str):
    index = [index]

name_map = {'agent': 'Agent', 'conc': 'Concentration', 'mech': 'Mechanism', 'prof': data.columns.values[0]}
ind_col_names = [name_map[s] for s in index]

index2 = []
for idx in ind_col_names:
    index2.append([s for s in data[idx]])
tuples = list(zip(*index2))

data2 = data.copy()
data2.index = pd.MultiIndex.from_tuples(tuples, names=index)

#################################
# keep these columns in dataframe
#################################

columns = None
# columns = []
if columns is None:
    columns = []
    
if isinstance(columns, str):
    columns = [columns]

keep_columns = set(columns)
all_columns = set(name_map.keys())

# drop these columns
drop_columns = list(all_columns - keep_columns)
drop_columns = [name_map[s] for s in drop_columns]

if len(drop_columns) > 0:
    data2 = data2.drop(drop_columns, inplace=False, axis=1)

if 84 == len(data2.columns):
    col_index = np.transpose([s.split(':') for s in data2.columns.values])
    data2.columns = pd.MultiIndex.from_tuples(list(zip(*col_index)), names=['System', 'Marker'])

data2
```
