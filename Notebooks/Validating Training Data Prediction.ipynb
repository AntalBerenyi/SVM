# Validating Training Data Prediction
In this Notebook we veryfy that the training data classes can be "retrieved" successfully by the trained model. 
The table below lists the traing data profiles and the prediction classes. A good prediction is indicated by Decision Value > 1 for the mechanism class.

```python
from scripts.svm_tools import SVM
preds, ptable = SVM().predict()
```

```python
import pandas as pd
p2 = []
for prof, mec in preds.idxmax(1).items():
    dv = ptable[(ptable.Profile == prof) & (ptable.Mechanism == mec)]['Decision Value'].values[0]
    a = [prof, mec, dv]
    p2.append(a)

p2 = pd.DataFrame(p2)
p2.sort_values(by=[1,0]).reset
```
